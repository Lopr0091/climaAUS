## Entendimiento del negocio

Australia es un país enorme, con una geografía diversa y uno de los climas más secos del mundo. Aunque gran parte del territorio es árido, sobre todo en el centro, existen regiones con condiciones óptimas para la agricultura, especialmente en el sudeste y suroeste, donde el clima es más templado y hay mayor disponibilidad de agua.
Estas variaciones climáticas y geográficas hacen de la planificación agrícola un verdadero desafío, ya que sembrar en el lugar o momento incorrecto puede generar pérdidas significativas. Ante esto, el sector agrícola necesita herramientas que permitan tomar decisiones más precisas sobre qué, dónde y cuándo cultivar, considerando variables como temperatura, lluvias, lugar y estacionalidad.
Este proyecto busca entregar valor mediante el análisis de datos meteorológicos históricos para predecir condiciones adecuadas de siembra, facilitando una agricultura más eficiente, sostenible y adaptada al desafiante entorno australiano.


Objetivos claves
- •	Optimización de la toma de decisiones por medio del análisis de datos meteorológicos y características de los principales cultivos de Australia de modo de mejorar el rendimiento y eficiencia.
- •	Identificación de las condiciones ideales para la siembra de distintos cultivos de acuerdo con el lugar, época y características climáticas
- •	Reducir el riesgo de tomar malas medidas y decisiones de siembra gracias a predicciones que mitiguen los efectos adversos del clima australiano.
